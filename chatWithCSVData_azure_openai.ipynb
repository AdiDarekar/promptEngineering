```python
import json
import pandas as pd
from google.colab import userdata
from openai import AzureOpenAI
#from dotenv import dotenv_values
```

```python
azure_openai_api_key = userdata.get('newName')
```

```python
client = AzureOpenAI(
    api_key=azure_openai_api_key,
    api_version="2023-05-15",
    azure_endpoint="https://oaid.openai.azure.com/"
)
```

```python
#df = pd.read_csv('news_short.csv')
df1 =  pd.read_excel('TCS.xlsx',  'Profit & Loss')
df2 =  pd.read_excel('TCS.xlsx',  'Quarters')
```

```python
df2.head(2)
```

```python
query = 'what is the Net Profit?'
context = df2.head().to_json(orient='records')
```

```python
response = client.chat.completions.create(
    model = 'gpt35oaiD',
    messages=[
        {"role":"system", "content":"You are a helpful assistant"},
        {"role":"user", "content":"Context: " + context + "\n\n Query: " + query}
    ]
)
```
